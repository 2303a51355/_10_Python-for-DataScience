```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/daily_temperature.csv")
print(df.head())
plt.figure(figsize=(10, 5))
plt.plot(df["Day"], df["Temperature"], marker="o", linestyle="-", color="b")
plt.title("Daily Temperature Changes Over a Month")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
# Add grid lines
plt.grid(True, linestyle="--", alpha=0.7)
# Show the plot
plt.show()
# Find hottest and coldest day
hottest_day = df.loc[df["Temperature"].idxmax()]
coldest_day = df.loc[df["Temperature"].idxmin()]
print(f" Hottest Day: Day {hottest_day['Day']} with {hottest_day['Temperature']}C")
print(f" Coldest Day: Day {coldest_day['Day']} with {coldest_day['Temperature']}C")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/sales_data.csv")

plt.figure(figsize=(8, 5))
plt.plot(df["Quarter"], df["Sales"], marker="o", linestyle="-", color="b", label="Sales")
peak = df.loc[df["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color="red", s=100, label="Peak Sales")
plt.title("Quarterly Sales Performance")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.show()
print(f" Peak Sales: {peak['Sales']} in {peak['Quarter']}")
```
